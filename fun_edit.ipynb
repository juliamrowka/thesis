```python
#preprocesory
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import Normalizer

import pandas as pd
```

```python
from sklearn import datasets
iris = datasets.load_iris()

print(iris.data)
print(type(iris.data))
print(type(iris))
iris.feature_names
```

```python
scaler = StandardScaler()
scaler.fit(iris.data)
newdata = scaler.transform(iris.data)
newdata
```

```python
scaler2 = Binarizer()
scaler2.fit(newdata)
newdata2=scaler2.transform(newdata)
newdata2
```

```python
from sklearn.pipeline import Pipeline


pipe=Pipeline([('std',StandardScaler()),('bin',Binarizer())])
pipe.fit(iris.data)
newdata3=pipe.transform(iris.data)

print(Pipeline)

newdata3
```

```python
from sklearn.compose import ColumnTransformer

ct= ColumnTransformer([('std',StandardScaler(),[0]),('minmax',MinMaxScaler(feature_range=(5,10)),[2])], remainder="passthrough")

ct.fit(iris.data)
res_ct=ct.transform(iris.data)

res_ct
```

```python
ct0= ColumnTransformer([('std',StandardScaler(),[0]), ('norm', Normalizer(), [1]), ('minmax',MinMaxScaler(feature_range=(5,10)),[2])], remainder="passthrough")
ct0.fit(iris.data)
res_ct0=ct0.transform(iris.data)

res_ct0
```

```python
ct2= ColumnTransformer([], remainder="passthrough")

ct2.transformers.append(('std',StandardScaler(),[0]))

print(type(ct2.transformers))
print(ct2)

ct2.fit(iris.data)
res_ct2=ct2.transform(iris.data)

res_ct2
```

```python
pipe2=Pipeline([('ct',ct)])
print(pipe2.steps)
pipe2.steps.append(['bin', Binarizer()])
print(pipe2.steps)

pipe2.fit(iris.data)
res_pipe2=pipe2.transform(iris.data)
res_pipe2
```

```python
from sklearn.decomposition import PCA
import numpy as np
pipe3=Pipeline([('std',StandardScaler()),('pca',PCA(n_components=3))])

pipe3.fit(iris.data)
res_pipe3=pipe3.transform(iris.data)
res_pipe3
```

```python
np.var(res_pipe3,axis=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
X_train,X_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2,random_state=0)

pipe4=Pipeline([('std',StandardScaler()),('pca',PCA(n_components=3)),('tree',DecisionTreeClassifier())])

pipe4.fit(X_train,y_train)
res=pipe4.predict(X_train)
np.transpose(np.array([y_train,res]))


```

```python
pipe4.score(X_test,y_test)
```

```python
pipe4.set_params(pca__n_components=4)
pipe4.fit(X_train,y_train)
pipe4.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(pipe4,X_train,y_train,cv=10)
```

```python
print(f"Średnia {scores.mean()}, odchylenie standardowe {scores.std()}")
```
